This project embarks on an in-depth Exploratory Data Analysis (EDA) of Airbnb listings in New York City for the year 2019. The primary objective is to uncover patterns, trends, and key characteristics within the bustling NYC Airbnb market. By analyzing various attributes of the listings, hosts, and review patterns, we aim to gain a comprehensive understanding that can benefit both potential hosts in optimizing their offerings and travelers in making informed booking decisions. The dataset serves as a rich source of information, encompassing details such as property specifications, host engagement, geographical spread, pricing strategies, and guest feedback.

The dataset under scrutiny is substantial, comprising 48,895 individual Airbnb listings, each described by 16 distinct features. These features are crucial for a holistic analysis and include identifiers like id and host_id, descriptive information such as name and host_name, geographical coordinates (latitude, longitude), and categorical attributes like neighbourhood_group, neighbourhood, and room_type. Furthermore, numerical features like price, minimum_nights, number_of_reviews, reviews_per_month, calculated_host_listings_count, and availability_365 provide quantitative insights into the listings' operational aspects and popularity.

A critical first step in any EDA is assessing data quality. Our initial examination revealed the absence of any duplicate rows, which is a positive indicator for the reliability of the dataset's unique entries. However, several columns exhibited missing values. Specifically, name had 16 missing entries, and host_name had 21. More significantly, last_review and reviews_per_month both showed 10,052 missing values. The high count in these review-related columns suggests that a substantial portion of listings might not have received any reviews yet, or their review data was not recorded. Addressing these missing values will be a key part of the data wrangling process, likely involving imputation or strategic handling to preserve the integrity of subsequent analyses.

The geographical distribution of Airbnb listings is a paramount aspect of this analysis. The listings are spread across five major neighbourhood_groups within NYC: Brooklyn, Manhattan, Queens, Staten Island, and Bronx. Visualizations confirmed that Manhattan and Brooklyn dominate the Airbnb market, hosting the vast majority of listings. This concentration is likely due to their status as major tourist attractions, business hubs, and densely populated residential areas. Understanding this distribution is vital for identifying prime locations and understanding supply-demand dynamics across different parts of the city.

The dataset categorizes accommodations into three room_types: 'Entire home/apt', 'Private room', and 'Shared room'. Analysis revealed that 'Entire home/apt' and 'Private room' are by far the most prevalent options, significantly outnumbering 'Shared room' listings. This trend suggests a strong preference among Airbnb users for privacy and independent living spaces, whether it's an entire unit or a private room within a shared dwelling. This insight can guide hosts in tailoring their offerings to meet prevailing guest demands.

The price column displays a considerable range, from listings offered at $0 to those reaching $$0 to those reaching $10,000. The presence of listings priced at $0 warrants further investigation, as these could represent data entry errors, promotional offers, or unique arrangements not explicitly detailed. On average, prices tend to be higher for 'Entire home/apt' compared to 'Private room' and 'Shared room', which is an expected finding reflecting the level of privacy and amenities offered. Furthermore, the average price varies significantly across different neighbourhood_groups, with Manhattan typically commanding higher prices due to its prime location and desirability.

The dataset provides glimpses into host activity through host_name and calculated_host_listings_count. Some hosts manage a considerable number of properties, with "Michael" identified as the host with the most listings (417). This indicates the presence of professional hosts or property management entities operating on the platform. The number_of_reviews and reviews_per_month columns serve as proxies for listing popularity and host engagement. Listings with higher review counts generally indicate established and well-regarded properties. The absence of reviews for a significant portion of listings could imply newer properties, less active listings, or those that have not yet garnered guest feedback. Overall, the initial exploration sets a strong foundation for deeper analysis into the factors influencing Airbnb success in NYC.
